```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import the data
df = pd.read_csv('msd2.csv', index_col=0)
```

```python
print(df.to_string())
```

```python
#reverse the df so it is in chronological order
dfr = df.iloc[len(df.index)::-1]
```

```python
ax = sns.boxplot(
    data=dfr,
    x='album', y='danceability',
    orient='v',
    hue='category',
    palette='inferno'
)
```
